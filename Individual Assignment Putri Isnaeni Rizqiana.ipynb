```python
#2. Top 3 Movies Net Profit in 2013
imdb_2013 = imdb[imdb['title_year'] == 2013].sort_values(by=['net_USD'],ascending=False).head(3)
imdb_2013
```

```python
#3. Top 3 Movies Gross Profit in 2009
imdb_2009 = imdb[imdb['title_year'] == 2009].sort_values(by=['gross_USD'],ascending=False).head(3)
imdb_2009
```

```python
#4. 
imdb_nonenglish = len(imdb[imdb['language'] != 'English'])
imdb_nonenglish
```

```python
#5. Top 3 Value aside English Language

imdb_nonenglish = (imdb[imdb['language'] != 'English'])
imdb_nonenglish.language.value_counts().head(3)
```

```python
#6. Top 3 countries with highest net profit
imdb.groupby('country').sum().sort_values(by='net_USD',ascending=False).head(3)
```

```python
#7. Average budget 2013 movies 
imdb_2013 = imdb[imdb['title_year'] == 2013]
imdb_2013mean = imdb_2013['budget_USD'].mean()
imdb_2013mean
```

```python
#8. Average budget 2014 movies 
imdb_2014 = imdb[imdb['title_year'] == 2014]
imdb_2014mean = imdb_2014['budget_USD'].mean()
imdb_2014mean
```

```python
#9. Higher average budget between 2013 and 2014
if imdb_2013mean > imdb_2014mean:
    print("Average Budget movie in 2013 is higher than 2014")
else:
    print("Average Budget movie in 2014 is higher than 2013")
```
